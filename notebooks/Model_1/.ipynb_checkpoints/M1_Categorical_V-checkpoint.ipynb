### Numerical Variables distribution Visuals

'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History' 

In this notebook, I look at numerical variable distributions using box plots and histograms. 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns  
df = pd.read_csv("Model_1/data.csv") 
```

```python
# Outliers, numerical
sns.set_context(context = "talk")
plt.figure(figsize=(16,5))
plt.figure(1,figsize=(20,10))
plt.subplot(121)
sns.distplot(df.ApplicantIncome)
plt.subplot(122)
sns.boxplot(x=df['ApplicantIncome'], whis=20);
plt.show()
```

```python
#Log scale 
df['ApplicantIncome_Log'] = np.log(df['ApplicantIncome']) 
```

```python
# Outliers, numerical
sns.set_context(context = "talk")
plt.figure(figsize=(16,5))
plt.figure(1,figsize=(20,10))
plt.subplot(121)
sns.distplot(df.ApplicantIncome_Log)
plt.subplot(122)
sns.boxplot(x=df['ApplicantIncome_Log'], whis=20);
plt.show()
```

```python
plt.figure(1,figsize=(20,10))
plt.subplot(121)
sns.distplot(df.CoapplicantIncome)
plt.subplot(122)
sns.boxplot(x=df['CoapplicantIncome'], whis=20);
plt.show()
```

```python
plt.figure(1,figsize=(20,10))
plt.subplot(121)
sns.distplot(df.LoanAmount)
plt.subplot(122)
sns.boxplot(x=df['LoanAmount'], whis=20);
plt.show()
```

```python
plt.figure(1,figsize=(20,10))
plt.subplot(121)
sns.distplot(df.Loan_Amount_Term)
plt.subplot(122)
sns.boxplot(x=df['Loan_Amount_Term'], whis=20);
plt.show()
```

```python
plt.figure(1,figsize=(20,10))
plt.subplot(121)
sns.distplot(df.Credit_History)
plt.subplot(122)
sns.boxplot(x=df['Credit_History'], whis=20);
plt.show()
```
