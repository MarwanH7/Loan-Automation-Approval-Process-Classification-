### Numerical Variables distribution Visuals

'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History' 

In this notebook, I look at numerical variable distributions using box plots and histograms. 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns  
df = pd.read_csv("data.csv") 
```

## ApplicantIncome Distribution 

```python
# Outliers, numerical
sns.set_context(context = "talk")
plt.figure(figsize=(16,5))
plt.figure(1,figsize=(20,10))
plt.subplot(121)
sns.distplot(df.ApplicantIncome)
plt.subplot(122)
sns.boxplot(x=df['ApplicantIncome'], whis=20);
plt.show()
```

## ApplicantIncome Distribution Log Scaled

```python
#Log scale 
df['ApplicantIncome_Log'] = np.log(df['ApplicantIncome']) 
```

```python
# Outliers, numerical
sns.set_context(context = "talk")
plt.figure(figsize=(16,5))
plt.subplot(121)
sns.distplot(df.ApplicantIncome_Log)
plt.subplot(122)
sns.boxplot(x=df['ApplicantIncome_Log'], whis=20);
plt.show()
```

## CoapplicantIncome Distribution

```python
df['CoapplicantIncome'] = df['CoapplicantIncome'].replace(0, np.NaN) # replace 0's with NaNs
df = df.dropna()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(121)
sns.distplot(df.CoapplicantIncome)
plt.subplot(122)
sns.boxplot(x=df['CoapplicantIncome'], whis=20);
plt.show()
```

```python
#Log scale 
df['CoapplicantIncome_log'] = np.log(df['CoapplicantIncome']) 
```

```python
# Outliers, numerical
sns.set_context(context = "talk")
plt.figure(figsize=(16,5))
plt.figure(1,figsize=(20,10))
plt.subplot(121)
sns.distplot(df.CoapplicantIncome_log)
plt.subplot(122)
sns.boxplot(x=df['CoapplicantIncome_log'], whis=20);
plt.show()
```
